```python
#Answer to 1.4

#setting normal range for duration 
durains = jan['Duration'].astype('timedelta64[s]')
dura_q1 = jan['Duration'].astype('timedelta64[s]').quantile(.25)
dura_q3 = jan['Duration'].astype('timedelta64[s]').quantile(.75)
dura_IQR = dura_q3-dura_q1
if dura_q1-1.5*dura_IQR > 0:
    dura_lower = duration
else:
    dura_lower = np.amin(durains)
dura_upper = dura_q3+3*dura_IQR  
#for any values 3.0*IQR above third quartile or 1.5*IQR below quartile can be 
#suspected as outliers, but the 1.5*IQR below quartile is smaller than 0
#therefore, the range is between minmum value and 3.0*iqr above third quartile
durains = jan['Duration'].astype('timedelta64[s]')
dura_outlier = durains[(durains > dura_upper)|(durains < dura_lower)]
dura_num_outlier = dura_outlier.count()
dura_total = durains.count()
dura_ratio = dura_num_outlier/dura_total
print("*"*86)
print('{:0.2f}% of instances in‘{}’are noisy! (normal range is between {} and {}) \n{}'.format(dura_ratio*100,'Duration',dura_lower,dura_upper,'*'*84))
#replace values 
dura_mean = jan['Duration'].mean()
jan.loc[(durains > dura_upper)|(durains < dura_lower), 'Duration'] = dura_mean


#setting normal values for trip distance
jan_trip = jan['Trip_distance']
trip_q1 = jan_trip.quantile(.25)
trip_q3 = jan_trip.quantile(.75)
trip_IQR = trip_q3-trip_q1
if trip_q1 - 1.5*trip_IQR > 0:
    trip_lower = trip_q1
else:
    trip_lower = np.amin(jan_trip)
trip_upper = trip_q3+3*trip_IQR
#if the value is 3*IQR over third quartile, then it must be a outliers 
#and 1.5*IQR below first quartile is samller than 0, then i choose
#the minmum number to be the lower bound
trip_total = jan_trip.count()
trip_outlier = jan_trip[(jan_trip > trip_upper)|(jan_trip < trip_lower)]
trip_num_outlier = trip_outlier.count()
trip_ratio = trip_num_outlier/trip_total
print("*"*86)
print('{:0.2f}% of instances in‘{}’are noisy! (normal range is between {} and {:0.2f}) \n{}'.format(trip_ratio*100,'Trip distance',trip_lower,trip_upper,'*'*84))
#replac values
trip_mean = jan_trip.mean()
jan.loc[(jan_trip > trip_upper)|(jan_trip < trip_lower),'Trip_distance'] = trip_mean


##setting normal values for passanger count
jan_pass = jan['Passenger_count']
pass_q1 = jan_pass.quantile(.25)
pass_q3 = jan_pass.quantile(.75)
pass_IQR = pass_q3-pass_q1
pass_lower = 1
pass_upper = pass_q3+3*trip_IQR
#it is acceptable that 1 or more people in a cab 
#but it would be unusual if more than 3*IQR above third quartile
#in a cab, it seems like overload
pass_total = jan_pass.count()
pass_outlier = jan_pass[(jan_pass > pass_upper)|(jan_pass < pass_lower)]
pass_num_outlier = pass_outlier.count()
pass_ratio = pass_num_outlier/pass_total
print("*"*86)
print('{:0.2f}% of instances in‘{}’are noisy! (normal range is between {} and {:0.2f}) \n{}'.format(pass_ratio*100,'Passnager_count',pass_lower,pass_upper,'*'*84))
#replace 
pass_mean = jan_pass.mean()
jan.loc[(jan_pass > pass_upper)|(jan_pass < pass_lower),'Passenger_count'] = pass_mean
```

```python
#Answer to 2.1
new_jan = pd.read_csv("clean-january.csv",encoding = 'ISO-8859-1')
new_jan['pickup_datetime'] = pd.to_datetime(new_jan['pickup_datetime'],format='%d/%m/%y %H:%M')
new_jan['Weekend']=new_jan['pickup_datetime'].dt.dayofweek
week = new_jan.loc[:,'Weekend']
total = week.count()
weekends = week[week >= 5].count()
new_jan.loc[week < 5,'Weekend'] = 0
new_jan.loc[week >=5,'Weekend'] = 1

print("*"*80)
print("% of weekend trips={:0.2f}".format(weekends/total*100))
print("*"*80)
```
